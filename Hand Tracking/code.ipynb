```python
import cv2 as cv
import mediapipe as mp
import time
```

```python
mpHands = mp.solutions.hands
hands = mpHands.Hands()
mpDraw = mp.solutions.drawing_utils
```

```python
cap = cv.VideoCapture(0)

pTime = 0
cTime = 0

while True:
    isTrue, frame = cap.read()
    imgOG = cv.resize(frame, (640,480))


    imgRGB = cv.cvtColor(imgOG, cv.COLOR_BGR2RGB)
    results = hands.process(imgRGB)
    if results.multi_hand_landmarks:
        for handlmrk in results.multi_hand_landmarks:
            for id, lm in enumerate(handlmrk.landmark):
                # print(id,lm)
                h, w, c = imgOG.shape
                cx, cy = int(lm.x*w), int(lm.y*h)
                # print(id, cx, cy)
                cv.circle(imgOG, (cx, cy), 5, (0, 0, 255), cv.FILLED)
            mpDraw.draw_landmarks(imgOG, handlmrk, mpHands.HAND_CONNECTIONS)

    cTime = time.time()
    fps = 1/(cTime-pTime)
    pTime = cTime
    cv.putText(imgOG, str(int(fps)), (10, 70), cv.FONT_HERSHEY_COMPLEX, 3, (255,0,0), 3)

    cv.imshow("webCam", imgOG)
    if cv.waitKey(1) & 0xFF == ord('x'):
        break

cap.release()
cv.destroyAllWindows()
```
